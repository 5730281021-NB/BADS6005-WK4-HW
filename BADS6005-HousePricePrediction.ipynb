```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.gridspec as gridspec 
from scipy import stats
import seaborn as sns
sns.set_style('whitegrid')

import warnings
warnings.simplefilter(action='ignore')

import pickle

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score, GridSearchCV, KFold, RandomizedSearchCV, train_test_split
import math
import sklearn.model_selection as ms
import sklearn.metrics as sklm
```

```python
x_train = pd.read_csv('train-2.csv')
train2 = pd.read_csv('train.csv')
# train.head(1460)
x_test = pd.read_csv('test-2.csv')
y_train = train2['SalePrice']
print(x_train)
```

```python
plt.figure(figsize = (15,5),dpi=100)
plt.subplot(131)
plt.scatter(x_train['LotArea'],y_train,alpha=0.5,c='RED')
plt.xlabel("Lot Area(X1)")
plt.ylabel("Price (Y)")

plt.subplot(132)
plt.scatter(x_train['1stFlrSF'],y_train,alpha=0.5)
plt.xlabel("First Floor square feet(X2)")
plt.ylabel("Price (Y)")

plt.subplot(133)
plt.scatter(x_train['2ndFlrSF'],y_train,alpha=0.5,c='Green')
plt.xlabel("Second Floor square feet(X3)")
plt.ylabel("Price (Y)")
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x_train,y_train)
print(np.round(lin_reg.intercept_,2),np.round(lin_reg.coef_,2))
```

```python
result = lin_reg.predict(x_test)
print(result)
```
